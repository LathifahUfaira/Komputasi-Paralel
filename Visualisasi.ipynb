```python
import matplotlib.pyplot as plt

process_counts = [1, 2, 4, 8, 16, 32, 64]
speedups = [1.6586, 2.0056, 1.6628, 2.0293, 1.6263, 1.9720, 1.4653]

plt.figure(figsize=(10, 6))
plt.plot(process_counts, speedups, 'bo-', label='Speedup Aktual')
plt.plot(process_counts, process_counts, 'r--', label='Speedup Ideal')

plt.xscale('log', base=2)
plt.xticks(process_counts, process_counts)

plt.xlabel('Jumlah process')
plt.ylabel('Speedup')

plt.title('Speedup vs Jumlah process')
plt.legend()
plt.grid(True)
plt.savefig('speedup_analysis.png')
plt.show()

#perbaiki kodenya ni, dimana salahnya
```

```python
import matplotlib.pyplot as plt

# Data
samples = ["100.000", "500.000", "1.000.000", "5.000.000", "10.000.000"]
waktu_eksekusi = [0.1339, 0.3868, 0.7256, 3.4311, 6.4955]

# Membuat grafik
plt.figure(figsize=(10, 5))

plt.plot(
    samples,
    waktu_eksekusi,
    marker='o',
    linewidth=2
)

# Label dan judul
plt.xlabel("Samples")
plt.ylabel("Waktu Eksekusi (detik)")
plt.title("Waktu Eksekusi vs Ukuran Sample")

# Grid
plt.grid(True)

# Simpan gambar
plt.savefig("waktu_eksekusi_vs_sample.png")

# Tampilkan grafik
plt.show()
```

```python
import matplotlib.pyplot as plt

speed_up = ["1.6586", "2.0056", "1.6628", "2.0293", "1.6263", "1.9720", "1.4653"]
efficiency = [1.6586, 1.0028, 0.4157, 0.2536, 0.1016, 0.0616, 0.0228]

plt.figure(figsize=(10,5))
plt.plot(speed_up, efficiency, marker='o')
plt.xticks(speed_up)

plt.xlabel("Speed UP")
plt.ylabel("Efficency")

plt.title("Speed Up vs Efficiency")
plt.savefig('speedup_vs_efficiency.png')

plt.grid()

plt.show()
```
